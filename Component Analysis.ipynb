```python
import networkx as nx
import pandas
import numpy as np
import json
%pylab inline
```

```python
with open("data/matrixSeries.json") as f:
    array_series = json.load(f)
graph_series = [nx.Graph(np.array(array)) for array in array_series]
```

```python
largest_component = [len(nx.connected_components(graph)[0]) for graph in graph_series]
largest_component = pandas.Series(largest_component)
largest_component.plot(title="Scenario 3: Largest component Size")
```
